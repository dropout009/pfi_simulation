```python
import pandas as pd
import numpy as np
import scipy as sp

from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor
from xgboost import XGBRFRegressor

import shap

import matplotlib.pyplot as plt
import seaborn as sns
import plotly_express as px
from plotnine import *

from sklearn.ensemble import ExtraTreesClassifier
plt.rcParams['figure.dpi'] = -0

import warnings
warnings.simplefilter('ignore')
np.random.seed(42)
```

```python
N = 1000
D = 9

X = np.random.uniform(0, 1, size=[N, D])
beta = np.array([1, 2, 3, 0, 0, 0, 0, 0, 0])
e = np.random.normal(0, 1, N)

y = X @ beta + e
```

```python
model = RandomForestRegressor(n_estimators=500, min_samples_leaf=1, n_jobs=-1, random_state=42)
model.fit(X, y)

pfi = permutation_importance(model, X, y, n_repeats=10, n_jobs=-1, random_state=42)
```

```python
df_pfi = pd.DataFrame(pfi['importances'])
df_pfi['variable'] = [f'x{d}' for d in range(1, D+1)]
df_pfi = df_pfi.melt(id_vars='variable', var_name='trial', value_name='importance')
```

```python
def create_data(N, D):
    X = np.random.uniform(0, 1, size=[N, D])
    beta = np.zeros(D)
    beta[0:3] = np.array([1, 2, 3])
    e = np.random.normal(0, 1, N)

    y = X @ beta + e
    return X, y

def train_rf(X, y):
    model = RandomForestRegressor(n_estimators=500, min_samples_leaf=1, n_jobs=-1, random_state=42)
    model.fit(X, y)
    
    return model

def calc_pfi(model, X, y, r):
    pfi = permutation_importance(model, X, y, n_repeats=5, n_jobs=-1, random_state=42)
    
    df = pd.DataFrame(pfi['importances_mean'])
    df['variable'] = [f'x{d}' for d in range(1, D+1)]
    df['round'] = r
    
    return df
```

```python
N = 1000
D = 9
R = 10

dfs = []
for r in range(R):
    print(r)
    
    X, y = create_data(N, D)
    model_trained = train_rf(X, y)
    tmp_df = calc_pfi(model_trained, X, y, r)
    dfs.append(tmp_df)

df = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
N = 1000
D = 9
R = 10
X, y = create_data(N, D) 
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42)

model = LGBMRegressor(n_estimators=10000, learning_rate=0.01, n_jobs=-1, random_state=42)
model.fit(X_train, y_train, eval_set=[(X_val, y_val)], early_stopping_rounds=50, verbose=100)
```

```python
def create_simulation_data(N, D, rho):
    mean = np.zeros(D)
    cov = np.eye(D)
    cov[0, 1] = rho
    cov[1, 0] = rho
    
    X = np.random.multivariate_normal(mean, cov, size=N)
    
    beta = np.zeros(D)
    beta[0] = 1
    e = np.random.normal(0, 1, N)
    
    y = X @ beta + e
    
    return X, y

X, y = create_simulation_data(N, D, 0.9)
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=42)

model = LGBMRegressor(n_estimators=10000, learning_rate=0.01, n_jobs=-1, random_state=42)
model.fit(X_train, y_train, eval_set=[(X_val, y_val)], early_stopping_rounds=50, verbose=100)

S_intervention = shap.TreeExplainer(model, X_val).shap_values(X_val)
S_tree = shap.TreeExplainer(model, feature_perturbation='tree_path_dependent').shap_values(X_val)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2, 1), dpi=100)
shap.summary_plot(S_tree, X_val, show=False)
fig.tight_layout()
fig.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2, 1), dpi=100)
shap.summary_plot(S_intervention, X_val, ax=ax, show=False)
fig.tight_layout()
fig.show()
```
